Avant de commencer à analyser des données, il est important de vérifier que les données sont effectivement disponibles. En vérifiant le code de statut, vous pouvez vous assurer que la requête vers le serveur s'est bien déroulée et que vous avez reçu une réponse

Dans les définitions de fonctions, le texte entre parenthèses après le nom de la fonction est appelé un **paramètre**. Les paramètres sont utilisés pour passer des valeurs à une fonction afin qu'elle puisse les utiliser dans son traitement.

Dans le cas ci-dessous de la fonction **GetHTML(url)**, le paramètre **url** est utilisé pour spécifier l'URL du site web à partir duquel nous voulons obtenir le contenu HTML. Lorsque nous appelons cette fonction, nous devons fournir une valeur pour **url** afin qu'elle puisse exécuter sa logique et récupérer le contenu HTML de cette URL spécifique.

```python
def GetHTML(url) :
    """
    Needed : url website
    Return : HTML text
    """
    response = requests.get(url)

    if response.status_code == 200 :
        print("Response OK, continue.")
    else :
        print("Access impossible.")
    
    return response.text
```

Dans le cas ci-dessous de la fonction **GetDataframe(html)**, le paramètre **page_medalist** est utilisé pour spécifier le contenu HTML à partir duquel nous voulons extraire les données pour créer un DataFrame. Lorsque nous appelons cette fonction, nous devons fournir le contenu HTML que nous avons récupéré précédemment (par exemple, en appelant GetHTML()).

La raison d'avoir la fonction ci dessous est de modulariser votre code.\
En encapsulant le processus de création du DataFrame dans une fonction distincte, vous pouvez réutiliser cette fonction chaque fois que vous avez besoin de créer un DataFrame à partir des données d'une page web. Cela rend votre code plus lisible, plus modulaire et plus facile à maintenir.\
De plus, cela facilite le partage et la réutilisation du code dans d'autres parties de votre projet ou dans d'autres projets.

```python
def GetDataframe(page_medalist): 

    soup = BeautifulSoup(page_medalist,"html5lib") #parse
    
    # Recherche du bon tableau selon description
    tableau = soup.find("table", class_="simpletable", style="text-align:left;")
    
    # Récupération de la liste (find_all) de toutes les lignes
    rows = tableau.find_all("tr")
    
    # Création du tableau (futur dataframe)
    line_medalist = []
    
    # Sur chaque ligne (= athlète)
    for row in rows :
        describe_athlete = row.text.split() #récupère le texte
        cells = row.find_all("td") #récupère la liste (find_all) de toutes les cellules de la ligne
        if len(cells) == 7 : #si il y a 7 cellules alors c'est un athlète
            medal = cells[0].find('img')['title'][0]  #titre de la médaille
            country = cells[1].find('img')['title']  #pays de l'athlète
            athlete_name = cells[2].text.strip()  #nom de l'athlète
            games = cells[3].find('img')['title']  #jeux olympiques
            sport = cells[4].find('img')['title']  #sport
            event = cells[6].text.strip()  # evénement
    
            #print(cells)
        
            # Ajoute les informations à la liste des athlètes
            line_medalist.append({
                'medal': medal,
                'country': country,
                'athlete_name': athlete_name,
                'games': games,
                'sport': sport,
                'event': event
                })
        else : #sinon ce n'est pas un athlète, passe à la ligne suivante
            next
    medalist = pd.DataFrame(line_medalist)
    return medalist
```

```python
url = "http://olympanalyt.com/OlympAnalytics.php?param_pagetype=Medals&param_dbversion=&param_country=FRA&param_games=ALL&param_sport=ALL"
html_brut = GetHTML(url)
GetDataframe(html_brut)
```

```python
url2 = "http://olympanalyt.com/OlympAnalytics.php?param_pagetype=RefCountries&param_dbversion=&param_country=CIV&param_games=ALL&param_sport=ALL"
html2 = GetHTML(url2)
```

```python
soup2 = BeautifulSoup(html2,"html5lib") 
```

```python
# Recherche du bon tableau selon description
# Il faut mettre un "_" après class pour differencier avec class
tableau2 = soup2.find_all("table", class_="simpletable", style="text-align:center;")[1]


# Récupération de la liste (find_all) de toutes les lignes
rows2 = tableau2.find_all("tr")



# Création du tableau (futur dataframe)
country_jo = []



# Sur chaque ligne (= athlète)
for row2 in rows2 :
    cells2 = row2.find_all("td") #récupère la liste (find_all) de toutes les cellules de la ligne
    if len(cells2) == 5 and len(cells2[0].text.strip())!=0:
        Code = cells2[0].text.strip()
        Country2 = cells2[1].text.strip()
        Continent = cells2[2].text.strip()  
        First_participation = cells2[3].text.strip() 
        Last_participation = cells2[4].text.strip()

        #print(cells)
    
        # Ajoute les informations à la liste des athlètes
        country_jo.append({
            'Code': Code,
            'Country': Country2,
            'Continent': Continent,
            'First_participation': First_participation,
            'Last_participation': Last_participation
            })
    else : 
        next
```

```python
# Création dataframe
df_country = pd.DataFrame(country_jo)
df_country
```

**DEFINITION :**
- **Itération :** Processus consistant à répéter une suite d'instructions, que ce soit par l'utilisation d'une fonction récursive ou d'une boucle.

- **Boucle infinie :** Une boucle dont la condition de terminaison n'est jamais remplie, ce qui la fait s'exécuter indéfiniment.

```python
#le i est la lettre du compteur 
i = 0

#Zip permet de chercher plusieurs listes et on doit la determiner dans le for 
for code_pays,nom_pays in zip(df_country.Code,df_country.Country):
    
    url = "http://olympanalyt.com/OlympAnalytics.php?param_pagetype=Medals&param_dbversion=&param_country="+ code_pays +"&param_games=ALL&param_sport=ALL"
    html_brut = GetHTML(url)
    if i == 0:
        i += 1 #incrementation , on ajoute 1 pour creer un dataframe 
        df_final = GetDataframe(html_brut)
    else :
        df_final = pd.concat([df_final,GetDataframe(html_brut)],ignore_index = True) 
```
